## Step 1 - Import libraries

```python
# %pip install --upgrade scipy networkx
```

```python
import networkx as nx
import matplotlib.pyplot as plt
import seaborn as sns
import numpy as np
import statistics
import pandas as pd
from scipy.stats import entropy
```

 ## Step 2 - additions

#### Plotting the graph options

```python
options = {
    'node_color': 'black',
    'node_size': 40,
    'width': 2,
}
```

#### Function for computing the avarage

```python
def Average(lst):
    return sum(lst) / len(lst)
```

## Step 3 - Erdos graph

 with 64 nodes and 243 edges

```python
n = 64 # 64 nodes
m = 243  # 243 edges
seed = 20160  # seed random number generators for reproducibility
G_er = nx.gnm_random_graph(n, m, seed=seed)
nx.draw(G_er, pos=nx.circular_layout(G_er), **options)
plt.title("Erdos graph")
plt.savefig('Erdos_graph.png')
plt.show()
```

## Step 4 - MT graph

```python
G_mt = nx.read_adjlist('MT.csv', comments='#', delimiter=None, create_using=nx.Graph, nodetype=None, encoding='utf-8')
nx.draw(G_mt, **options)
plt.title("MT graph")
plt.savefig('MT_graph.png')
plt.show()
```

## Step 16 - Degree Histogram

### MT graph

```python
dh_mt = nx.degree_histogram(G_mt)
dh_mt.sort()
print('Degree Histogram of MT')
print(dh_mt)
```

### Erdos graph

```python
dh_er = nx.degree_histogram(G_er)
dh_er.sort()
print('Degree Histogram of Erdos')
print(dh_er)
```

```python
# MT
plt.bar(*np.unique(dh_mt, return_counts=True), label = 'MT')

# Erdos
plt.bar(*np.unique(dh_er, return_counts=True),alpha = 0.8, label = 'Erdos')

plt.title("Degree histogram")
plt.xlabel("Degree")
plt.ylabel("# of Nodes")
plt.legend()
plt.savefig('Degree_histogram.png')
plt.show()
```

## Step 17 - Ego Netwwork

### MT graph

```python
mt_degree = sorted(G_mt.degree, key=lambda x: x[1], reverse=True)
node_highest_degree_mt = mt_degree[0][0]
print(node_highest_degree_mt)
```

```python
ego_mt = nx.ego_graph(G_mt, node_highest_degree_mt)
nx.draw(ego_mt, **options)
plt.title("MT Ego graph")
plt.savefig('MT_Ego_graph.png')
plt.show()
```

### Erdos graph

```python
er_degree = sorted(G_er.degree, key=lambda x: x[1], reverse=True)
node_highest_degree_er = er_degree[0][0]
print(node_highest_degree_er)
```

```python
ego_er = nx.ego_graph(G_er, node_highest_degree_er)
nx.draw(ego_er, **options)
plt.title("Erdos Ego graph")
plt.savefig('Erdos_Ego_graph.png')
plt.show()
```
